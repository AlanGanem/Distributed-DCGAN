```python
import seaborn as sns
import pandas as pd
import json
```

```python
with open('exp_results.json', 'r') as f:
    data = json.load(f)
```

```python
tuple_data = []
for key in data:
    for value in data[key]:
        tuple_data.append((int(key), float(value)))
```

```python
data = pd.DataFrame(tuple_data, columns = ['n_procs', 'time(s)'])
data['speedup'] = data.query('n_procs == 1')['time(s)'].mean()/data['time(s)']
data['efficiency'] = data['speedup']/data['n_procs']
```

```python
sns.lineplot(x = data['n_procs'], y = data['time(s)'])
```

```python
sns.lineplot(x = data['n_procs'], y = data['efficiency'])
```

```python
sns.lineplot(x = data['n_procs'], y = data['speedup'])
```
